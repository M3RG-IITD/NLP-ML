```python
import pandas as pd
import ptable as pt
import numpy as np

name_ids = {}
name_ele = {}
for i in pd.read_csv('./all_component.csv').values:
    try:
        name_ids[i[1]] = int(i[0])
        name_ele[i[0]] = i[1]
    except:
        name_ids[i[1]] = None
        name_ele[i[0]] = None

def formula_to_idx(list1):
    return [name_ids[i] for i in list1]

def idx_to_formula(list2):
    return [name_ele[j] for j in [str(i) for i in list2]]
```

```python
## Read dataframe
my_df = df
# print(my_df.columns)
#Identify keys (columns) common to both the DataFrame and the dictionary
common_columns = set(my_df.columns[:-1]).intersection(name_ids.keys())

# Create a new dictionary for renaming columns using the common keys
column_rename_dict = {old_col: name_ids[old_col] for old_col in common_columns}

# Rename columns based on the new dictionary
my_df.rename(columns=column_rename_dict, inplace=True)
print(my_df.columns)
X_comp = my_df.values[:,:-1]

glass_id = [int(i) for i in (my_df.columns.to_list()[:-1])]

comp_x_desc, eles, count = pt.get_descriptor(idx_to_formula(glass_id))
n_x_desc = np.dot(X_comp,comp_x_desc.values)
n_x_desc = pd.DataFrame(n_x_desc,columns = comp_x_desc.columns)
n_x_desc[['O_per', 'NF_per', 'NM_per',]] /= n_x_desc[['O_per', 'NF_per', 'NM_per',]].values.sum(axis=1).reshape(-1,1)

X_descr = n_x_desc
pd.concat([X_descr, my_df.iloc[:,-1]], axis=1).to_csv('descriptor.csv', index=False)
```
